```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb_top_1000.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
cert = df['Certificate'].value_counts()
```

```python
plt.pie(cert[:5],labels=cert.index[:5],autopct='%1.2f%%')
plt.title('Certificate Based analysis')
```

# Observation: 
U- Universal certificates movie in Top 100 is taking the lead followed by A- Adult Rated movies then UA,R,PG-13 taking top 5 Spots respectively

```python
df['IMDB_Rating'].mean()
```

# The average rating of top 1000 movie is 7.949

```python
genre = df['Genre'].value_counts()
```

```python
plt.pie(genre[:5],labels=['Drama','Romance','Comedy','RomCom','Action'])
plt.title('Movies based on Genre')
```

# Drama is the most repeated genre in top 1000

```python
direct = df['Director'].value_counts()
```

```python
plt.bar(direct.index[:5],direct[:5])
plt.tight_layout()
plt.xticks(rotation=45)
plt.xlabel('Name of the director')
plt.ylabel('Number of Movies')
plt.title('Highest number of movie of same director')
```

# Oberservation :
Alfred Hitchcock has most movies in top 1000

```python
year = df['Released_Year'].value_counts()
```

```python
plt.pie(year[:5],labels=year.index[:5],autopct='%1.2f%%')
plt.title('Year wise distribution of Movies in top 1000')
```

# Observation:
There is a almost tie between the movies of 2004,2014,2016,2013,2009


```python
all_star = pd.concat((df['Star1'],df['Star2'],df['Star3'],df['Star4']))
```

```python
star = all_star.value_counts()
```

```python
plt.bar(star.index[:5],star[:5])
plt.xlabel('Name of Actor')
plt.ylabel('Number of movies')
plt.title('Actors Repeated in most movies')
```
